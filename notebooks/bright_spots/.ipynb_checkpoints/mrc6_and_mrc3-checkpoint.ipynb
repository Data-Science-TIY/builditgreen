```python
import pandas as pd
import psycopg2
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import psycopg2
import pandas.io.sql as pdsql
import plotly.plotly as py
from plotly.graph_objs import *
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
df_leed_2009 = pd.read_pickle('leed_2009_w_credits.pickle')
```

```python
df_leed_2009.columns[:20]
```

```python
mrc6 =df_leed_2009[(df_leed_2009['mrc6'] > 0) & (df_leed_2009['Country'] == 'US')]
```

```python
mrc6_by_state = mrc6.groupby('State').count()["ID"]
mrc6_by_state = mrc6_by_state.to_frame()
mrc6_by_state = mrc6_by_state.rename(columns={"ID": "Number of Projects"})
```

#MRc6 by State

```python
mrc6_by_state.plot(kind='bar')
```

#All LEED 2009 projects by state
For analysis, need to compare to the total population of builgins to see if there is a difference

```python
df_leed_2009_by_state = df_leed_2009.groupby('State').count()["ID"]
df_leed_2009_by_state = df_leed_2009_by_state.to_frame()
df_leed_2009_by_state = df_leed_2009_by_state.rename(columns={"ID": "Number of Projects"})
df_leed_2009_by_state.plot(kind='bar', figsize=(12,10))
```

```python
mrc6_by_owner = mrc6.groupby('OwnerTypes').count()["ID"]
mrc6_by_owner = mrc6_by_owner.to_frame()
mrc6_by_owner = mrc6_by_owner.rename(columns={"ID": "Number of Projects"})
mrc6_by_owner.plot(kind='bar')
```

```python
#For analysis we need to compare how this is different from the total population of projects
```

```python
df_leed_2009_by_owner = df_leed_2009.groupby('OwnerTypes').count()["ID"]
df_leed_2009_by_owner = df_leed_2009_by_owner.to_frame()
df_leed_2009_by_owner = df_leed_2009_by_owner.rename(columns={"ID": "Number of Projects"})
df_leed_2009_by_owner.plot(kind='bar')
```
